```python
def f(x):
    """
    input : x
    output : giá trị hàm 3x^2 - 4x -2
    """
    return 3*x*x - 4*x -2
```

```python
def fd(x):
    """
    input : x
    output : giá trị hàm 6x - 4 
    """
    return 6*x - 4 
```

```python
def gra_des(LR, x_start , loop ):
    """
    input : 
    LR - hằng số học 
    x_start - điểm bắt đầu chạy
    loop - số vòng
    output :
    tọa độ điểm gần giá trị cực tiểu của hàm số nhất
    """
    for i in range(loop):
        x_start = x_start- LR*fd(x_start)
        if abs(fd(x_start)) <=1e-6:
            break
    x_end = x_start
    return x_end , f(x_end)

        
        
    
```

```python
def list_gra_des(LR, x_start , loop ):
    X = [x_start]
    Y = [f(x_start)]
    """
    input : 
    LR - hằng số học 
    x_start - điểm bắt đầu chạy
    loop - số vòng
    output :
    tọa độ các điểm đi qua đến khi gần giá trị cực tiểu của hàm số nhất
    """
    for i in range(loop):
        x_start = x_start- LR*fd(x_start)
        if abs(fd(x_start)) <=1e-6:
            break
        X.append(x_start)
        Y.append(f(x_start))
    return X,Y
```

```python
X,Y = gra_des(0.01,-5,10000000)
print (X,Y)
X1 =[]
Y1=[]
X1,Y1= list_gra_des(0.01,-5 ,10000000)
```

```python

t1= np.linspace(-5,5,100)
plt.plot(t1,f(t1))
plt.title("$3x^2 -4x -2$")
plt.scatter(X1,Y1,color = 'green')
plt.scatter(X,Y,color='red')
plt.annotate("Global Minium " , xy = (X,Y), xytext = (0,10) , arrowprops = dict(facecolor = 'red', shrink = 1) )
plt.annotate("start " , xy = (X1[0],Y1[0]), xytext = (-4,90) , arrowprops = dict(facecolor = 'green', shrink = 1) )
plt.annotate("next step " , xy = (X1[1],Y1[1]), xytext = (-3.9,80) , arrowprops = dict(facecolor = 'green', shrink = 1) )
plt.savefig("output.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def F(x):
    """
    input:  x  
    output: trả về giá trị của hàm F(x)
    """
    return x * x - 5 * np.sin(x)

def f(x):
    """
    input: x
    output: trả về giá trị của đạo hàm F(x)
    """
    return 2 * x - 5 * np.cos(x)

def findX(LR,xold,loop):
    """
    input: 
         LR: learningRate (Hằng số học)
         xold: Điểm x ban đầu
         loop: số vòng lặp
    output:
        trả về giá trị cực tiểu của hàm số
    """
    xnew = xold
    for i in range(loop):
        xnew = xnew - LR * f(xnew)
        if f(xnew) <= 1e-6:
            break
    return xnew
    


X = findX(0.001,4,1000)
print("x = :",X)



"""
vẽ hình
"""
A = np.linspace(-5, 4,100)
plt.plot(A,F(A))
plt.title("$F(x) = x^2 - 5 \cdot sin(x)$")
plt.scatter(X,F(X),color='red')
plt.annotate("Global_Minimun", xy = (X,F(X)) , xytext = (0,3),arrowprops=dict(facecolor='red', shrink=1))
plt.savefig("output.png")
```
